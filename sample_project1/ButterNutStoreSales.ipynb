# EXERCISE REPLICATE THE SHEET BELOW USING PANDAS AND NUMPY

```python
import numpy as np 

sales = np.random.randint(1,5, size=(5,3))
sales
```

```python
import pandas as pd
sales_pd = pd.DataFrame(sales, index = ["m", "t","w","th","f"], columns = ["Almond", "Peanut","Cashew"])
sales_pd
```

```python
prices = np.array([10,8,12]) 
prices
```

```python
prices_pd = pd.DataFrame(prices.reshape(1,3), index=["Prices"], columns = ["Almond", "Peanut","Cashew"])
prices_pd
```

```python
sales_pd.shape , prices_pd.shape
```

```python
sales.shape , prices.shape
```

```python
total_sales = sales.dot(prices)
total_sales
```

```python
sales_pd["Total"] = total_sales
sales_pd
```
